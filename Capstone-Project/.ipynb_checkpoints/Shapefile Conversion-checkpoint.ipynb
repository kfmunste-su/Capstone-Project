```python
geometry['latitude']='default value'
geometry['longitude'] = 'default value'
geometry.head()
```

```python
# write new dataframes to csv file
# compresses data with parquet files. makes files smaller and opens more quickly, 
# even if the time savings isn't very noticebale it can take up less storage with large amounts of datasets

geometry_subset = geometry.sample(3).to_dict(orient='records')
for y in geometry_subset:
    print(y)
```

```python
geometry_dict = geometry.to_dict(orient='records')
for i in geometry_dict:
    value = str(i["geometry"]).strip('POINT').strip(' (').strip(')')
    space_index = value.index(' ')
    i['latitude'] = float(value[0:space_index])
    i['longitude'] = float(value[space_index+1:])
```

```python
geometry_new = pd.DataFrame.from_dict(geometry_dict)
geometry_new
```

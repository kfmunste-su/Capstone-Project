```python
# Google Maps API no longer free, but data science toolkit is. 
# Data science toolkit does not do reverse geocoding, so used their "Coordinates 2 Politics" API instead to get the city and state.

new_york_data = {}

for i in range(0, len(data["trashpoints"])):
    latitude = data["trashpoints"][i]["lat"]
    longitude = abs(data["trashpoints"][i]["long"])
    url = "http://www.datasciencetoolkit.org/coordinates2politics/" + str(latitude) + "%2c-" + str(longitude)
    response = requests.get(url)
    geodata = response.json()
    while True:
        try:
            if (len(geodata[0]["politics"]) == 5):
                if geodata[0]["politics"][4]["name"] == "New York":
                    new_york_data[i] = data["trashpoints"][i]
        except TypeError:
# Some lat and long is creating empty geo-location results.
# Switch to google maps API?
```
